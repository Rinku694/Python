Task 1 - There are three different tables as given above. Please make three dataframe in python
and save them as three .csv files. From Task 2 to Task 10, use the saved .csv files only.


```python
import pandas as pd  # To use pandas function in python we need to import the pandas.

# Creating Project DataFrame by adding data given in PDF.
project_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'],
    'Project': ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8',
                'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'],
    'Cost': [1002000, 2000000, 4500000, 5500000, None, 680000, 400000, 350000, None, 300000, 2000000, 1000000, 3000000, 200000],
    'Status': ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed',
               'Ongoing', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Finished']
}
project_df = pd.DataFrame(project_data)  # converting the above data into DataFrame.
project_df.to_csv('Project.csv', index=False)  # Saving the data in csv file and applying index = false to stop index count.

pd.DataFrame(project_df)
```

```python
# Creating Employee DataFrame by adding data given in pdf.
employee_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],
    'Age': [25, 27, 29, 31, 30]
}
employee_df = pd.DataFrame(employee_data) # converting the above data into DataFrame.
employee_df.to_csv('Employee.csv', index=False)  # Saving the data in csv file and applying index = false to stop index count.
pd.DataFrame(employee_df)
```

```python
# Creating Seniority Level DataFrame by adding data given in pdf.
seniority_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Designation Level': [2, 2, 3, 2, 3]
}
seniority_df = pd.DataFrame(seniority_data) # converting the above data into DataFrame.
seniority_df.to_csv('Seniority.csv', index=False)  # Saving the data in csv file and applying index = false to stop index count.
pd.DataFrame(seniority_df)
```

Task 2 - The cost column in the dataframe “Project” has some missing values. Your task is to
compute these missing values. Replace the missing values by running average. You
should use the “For” loop for this task.

```python
# Loading Project DataFrame
project_df = pd.read_csv('Project.csv')

# Replace missing values using running average
running_sum = 0
count = 0
for i in range(len(project_df)):          # If cost has NaN value then the given formula will run
    if pd.isna(project_df.loc[i, 'Cost']):
        project_df.loc[i, 'Cost'] = running_sum / count     # running avarage = runing_sum/count
    else:
        running_sum += project_df.loc[i, 'Cost']  # If cost has value, it add previous row cost with current row cost.
        count += 1                                # And count will add by one.

project_df.to_csv('Project.csv', index=False)  # Saving the upated changes in csv file.
pd.DataFrame(project_df)
```

Task 3 - Split the name column in the Employee dataframe into two new columns “First Name”,
and “LastName” and remove the older “name” column.


```python
# Loading Employee DataFrame.
employee_df = pd.read_csv('Employee.csv')

# As per given question first we create two column by 'first_name' and 'last_name' and then breaking the name column with the help of str.split().
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(pat=' ', n=1, expand=True)

# After seperating the name column into 'firt_name' and 'last_name', Now name column is extrated, i will remove it by using drop function.
employee_df.drop(columns=['Name'], inplace=True)

# Saving the updated changes in csv file.
employee_df.to_csv('Employee.csv', index=False)

pd.DataFrame(employee_df)
```

Task 4 - Join all three dataframes in one single dataframe. Name it “Final”

```python
# As we already loaded the employee_df and project_df in above code.
# Now loading seniority_df data.
seniority_df = pd.read_csv('Seniority.csv')

# Using pd.merge, It can merge two table at a time, So merging employee table and seniority table by taking ID column common.
final_df = pd.merge(employee_df, seniority_df, on='ID')

# Now merging project table with final table by taking ID column common.
final_df = pd.merge(final_df, project_df, on='ID')

# Now we merge all three tales in one table which 'final_df'
#Saving the final table in csv file.
final_df.to_csv('Final.csv', index=False)

pd.DataFrame(final_df)
```

Task 5 - Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project 
cost only to employees who have finished the projects.


```python
# In above code we merge all three table in final table.
# Now, loading the final table.
final_df = pd.read_csv('Final.csv')

# Applying 'lambda function' in each rows to add 5% bonus only to employee who finished the projects.
# axis = 1 just to see function is applied row wise. 
final_df['Bonus'] = final_df.apply(lambda row: row['Cost'] * 0.05 if row['Status'] == 'Finished' else 0, axis=1)

# Saving the changes in final table into csv file.
final_df.to_csv('Final.csv', index=False)

pd.DataFrame(final_df)
```

Task 6 - Demote the designation level by 1, whose projects have status “fail”. Delete the
employees record whose designation level is above 4.


```python
# Now, no need to load the final data because its already loaded in previous code.
# Using LOC function to identify status = failed and designation level + 1.
final_df.loc[final_df['Status'] == 'Failed', 'Designation Level'] += 1

# Here we are filtering the row which has 'Designation Level' less then and equal to 4.
final_df = final_df[final_df['Designation Level'] <= 4]

# Saving the changes in final table into csv file.
final_df.to_csv('Final.csv', index=False)

pd.DataFrame(final_df)
```

Task 7 - Add “Mr.” and “Mrs.” to the first name column and drop the gender column.


```python
# Using Lambda & if an else function to add Mr. & Mrs. 
final_df['First Name'] = final_df.apply(
    lambda row: f"Mr. {row['First Name']}" if row['Gender'] == 'M' else f"Mrs. {row['First Name']}", axis=1)

# Droping Gender column because gender of name will identify by seening Mr. & Mrs. in first_name column.
final_df.drop(columns=['Gender'], inplace=True)

# Saving the changes in final table into csv file.
final_df.to_csv('Final.csv', index=False)

pd.DataFrame(final_df)
```

Task 8 - Promote designation level by 1 for the employees whose age is more than 29 years using
IF condition.


```python
# Using loc function identifying whoes age is greater then 29, decreasing 'Designation level' by -1.
# Because in pdf its mention if 'designation level' is low it mean the employee is at higher post & if its high it mean employee is at lowest post.
final_df.loc[final_df['Age'] > 29, 'Designation Level'] -= 1

# Saving the changes in final table into csv file.
final_df.to_csv('Final.csv', index=False)

pd.DataFrame(final_df)
```

Task - 9 Add the cost of all projects for each Employee and save it in new dataframe
“TotalProjCost” with three columns ID, First Name, and Total cost

```python
# Grouping ID and First Name to create unique id, Finding total cost of each employee by applying sum aggregation.
# Using reset_index to Convert the grouped data back into a standard DataFrame and rename the aggregated column.
total_proj_cost = final_df.groupby(['ID', 'First Name'])['Cost'].sum().reset_index(name='Total Cost')

# Saving the changes in TotalProjcost table into csv file.
total_proj_cost.to_csv('TotalProjCost.csv', index=False)

pd.DataFrame(total_proj_cost)
```

```python
# Flittering the final_df data column 'city' contain 'o' make it True else False
# All city row which contains 'o' in there name will be in data but a city which doent contains 'o' will gets deleted.
filtered_employees = final_df[final_df['City'].str.contains('o', case=False)]

# Printing 'filtered_employee
pd.DataFrame(filtered_employees)
```
